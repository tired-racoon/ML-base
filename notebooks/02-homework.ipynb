## Medium

```python
df['Age'] = 2024 - df['Year_Birth']
df.head()
```

```python
data = df[['Education','Marital_Status','Income','Kidhome','Teenhome','Recency','NumWebPurchases','NumStorePurchases','Age']]
data
```

```python
import seaborn as sns
```

```python
set(df['Education'])
```

```python
names = {'2n Cycle': 1, 'Basic': 2, 'Graduation': 3, 'Master': 4, 'PhD': 5}
def one_hot(txt):
    return names[txt]
```

```python
data['Edu'] = list(map(one_hot, data['Education'].values.tolist()))
data
```

```python
set(df['Marital_Status'])
```

```python
statuses = {'Absurd' : 1,
 'Alone' : 2,
 'Divorced' : 3,
 'Married': 4,
 'Single': 5,
 'Together': 6,
 'Widow': 7,
 'YOLO': 8}
def one_hot_stat(txt):
    return statuses[txt]
data['Stat'] = list(map(one_hot_stat, data['Marital_Status'].values.tolist()))
data
```

```python
data = data.drop(['Education', 'Marital_Status'], axis = 1)
data
```

```python
sns.pairplot(data)
```

Можно видеть зависимости между Income и Num*-колонками. Высокая положительная корреляция. Также можно видеть высокую отрицательную корреляцию между Income и Kidhome. В целом можно сделать вывод: чем больше доход, тем больше человек тратит, но чем больше у него детей, тем он начинает тратить на себя меньше.
